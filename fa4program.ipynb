```python
number = 1
while number <= 20:
    if number % 2 == 0:
        number += 1
        continue
    print(number)
    number += 1
```

```python
correct_password = "jarvisyee"
max_attempts = 3
attempts = 0

print("Welcome to the Password Checker!")
while attempts < max_attempts:
    password_attempt = input(f"Enter the password (Attempt {attempts + 1}/{max_attempts}): ")
    if password_attempt == correct_password:
        print("Access granted")
        break
    else:
        print("Incorrect password")
        attempts += 1
if attempts == max_attempts:
    print("Too many incorrect attempts, try again later")
```

```python
print("Number Filter Display:")
for number in range(1, 11):
    if number == 4:
        continue
    elif number == 6:
        pass
    elif number == 8:
        break
    print(number)
```
